```python
def plot_for(datetime_, dataset):
    CS1, CS2 = cfs.plot_h5_anomaly(datetime_, basemap=maps.north_america())
    bottom_left_textbox(CS1.ax, 
                    '500mb Geopotential Heights (gpm)\nValid {} UTC\nDataset: {}'.format(datetime_, dataset))
    bottom_right_textbox(CS1.ax, 'Graphic by Jim Tang (@wxmann)')
    return CS1, CS2
```
